```python
# Import matplotlib library to create graphs and charts

import matplotlib.pyplot as plt

# Create a figure with width = 10 and height = 5 for better visibility
plt.figure(figsize=(10,5))


# Count how many customers are in each state and plot it as a bar chart

customers['customer_state'].value_counts().plot(kind='bar')

# Set the title of the graph
plt.title("Customer Distribution by State")

# Label for X-axis (State names)
plt.xlabel("State")

# Label for Y-axis (Customer count)
plt.ylabel("Number of Customers")

# Display the graph
plt.show()
```
